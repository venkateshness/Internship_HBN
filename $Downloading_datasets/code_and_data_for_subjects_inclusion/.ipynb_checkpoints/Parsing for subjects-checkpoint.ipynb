```python
import ezodf
import csv

def read_ods(filename, sheet_no=0, header=0):
    tab = ezodf.opendoc(filename=filename).sheets[sheet_no]
    return pd.DataFrame({col[header+1].value:[x.value for x in col[header+1:]]
                         for col in tab.columns()})

df_dwi = read_ods('dwi-subject_list.ods',0,0)
df_freesurfer = read_ods('subjects_freesurfer.ods')
```

```python
def parse(df_to_use):
    list_to_pass = list()
    for i in range(len(df_to_use.values)):
         for j in range(len(list(l))):
            if list(l)[j][4:] in df_to_use.values[i]:
                list_to_pass.append(list(l)[j])
    return list_to_pass
```
